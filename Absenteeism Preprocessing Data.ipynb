```python
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
import matplotlib.pyplot as plt
sns.set()
```

```python
rdata=pd.read_csv('Absenteeism_data.csv')
rdata.head()
```

```python
rdata.isna().sum()
```

```python
data=rdata.copy()
data=data.drop({'ID'},axis=1)
```

```python
pd.unique(data['Reason for Absence'])
```

```python
reason_columns=(pd.get_dummies(data['Reason for Absence']))
```

```python
reason_columns['check']=reason_columns.sum(axis=1)
reason_columns=reason_columns.drop({'check',},axis=1)
reason_columns=pd.get_dummies(data['Reason for Absence'],drop_first=True)
reason_columns
```

```python
reason_grp1=reason_columns.loc[:,1:14].max(axis=1)
reason_grp2=reason_columns.loc[:,15:17].max(axis=1)
reason_grp3=reason_columns.loc[:,18:21].max(axis=1)
reason_grp4=reason_columns.loc[:,22:].max(axis=1)
```

```python
data=data.drop({'Reason for Absence'},axis=1)
```

```python
data=pd.concat([data,reason_grp1,reason_grp2,reason_grp3,reason_grp4], axis=1)
```

```python
column_names=['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'GRP1', 'GRP2', 'GRP3', 'GRP4']
```

```python
data.columns=column_names
data
```

```python
reordered_column=['GRP1', 'GRP2', 'GRP3', 'GRP4','Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
data=data[reordered_column]
data.head()
```

```python
type(data['Date'][0])
```

```python
data['Date']=pd.to_datetime(data['Date'],format='%d%m%Y')
```

```python
data['Date'][0].month
```

```python
list_month=[]
for i in range(700):
    list_month.append(data['Date'][i].month)
data['Month Value']= list_month
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
data['Day of Week']= data['Date'].apply(date_to_weekday)
data.tail()
```

```python
cpdata=data.copy()
```

```python
cpdata=cpdata.drop({'Date'},axis=1)
cpdata.tail()
```

```python
cpdata.columns
```

```python
new_columns=['GRP1', 'GRP2', 'GRP3', 'GRP4','Month Value', 'Day of Week','Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average', 'Body Mass Index',
       'Education', 'Children', 'Pets', 'Absenteeism Time in Hours',
       ]
```

```python
cpdata=cpdata[new_columns]
cpdata.tail()
```

```python
cpdata['Education']=cpdata['Education'].map({1:0,2:1,3:1,4:1})
cpdata.tail()
```

```python
prodata=cpdata.copy()

```
